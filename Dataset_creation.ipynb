```python
import numpy as np
import pandas as pd

def generate_permutations(constant_values, num_permutations, num_items):
    all_permutations = []
    for i in range(num_permutations):
        # Generate a random permutation of the remaining items
        remaining_items = [x for x in num_items if x not in constant_values]
        perm = np.random.permutation(remaining_items).tolist()

        # Insert the constant values at their original indices
        full_perm = []
        j = 0
        for k in num_items:
            if k in constant_values:
                full_perm.append(constant_values[j])
                j += 1
            else:
                full_perm.append(perm.pop(0))
        all_permutations.append(full_perm)
    return all_permutations


constant_values = [4,21,9,7,11,23]  # indices start at 0
# []
num_permutations = 40
num_items = [6, 16, 4, 20, 21, 1, 5, 2, 8, 9, 3, 7, 19, 18, 17, 11, 15, 13, 25, 24, 14, 12, 10, 22, 23]
permutations = generate_permutations(constant_values, num_permutations, num_items)
pd.DataFrame(permutations).to_csv('1.csv')
```

```python
import pandas as pd
import numpy as np
from scipy.stats import kendalltau, spearmanr
from sklearn.decomposition import PCA

df = final_df

# Method 1: Borda Count
borda_rankings = df.sum().sort_values()

# Method 2: Kendall's Tau
kendall_correlation_matrix = df.corr(method='kendall')
average_kendall_rankings = kendall_correlation_matrix.mean().sort_values()

# Method 3: Spearman's Rank Correlation
spearman_correlation_matrix, _ = spearmanr(df)
average_spearman_rankings = pd.Series(spearman_correlation_matrix.mean(axis=0), index=df.columns).sort_values()

# Print the results
print("Borda Count Rankings:")
print(borda_rankings)

print("\nAverage Kendall's Tau Rankings:")
print(average_kendall_rankings)

print("\nAverage Spearman's Rank Correlation Rankings:")
print(average_spearman_rankings)

```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Set up the subplots
fig, axes = plt.subplots(nrows=5, ncols=5, figsize=(15, 15))

# Flatten the axes for easier iteration
axes = axes.flatten()

# Iterate over the columns of the dataframe and plot a distribution plot for each column
for i, col in enumerate(df.columns):
    sns.histplot(data=df, x=col, ax=axes[i], kde=True)

    # Add y-label count to leftmost figures only
    if i % 5 == 0:
        axes[i].set_ylabel('Count')
    else:
        axes[i].set_ylabel('')

    axes[i].set_xlabel(col)

# Hide any unused subplots
for i in range(len(df.columns), len(axes)):
    axes[i].axis('off')

# Add a title to the figure
fig.suptitle('Rank Assignment Distribution by The Annotators', fontsize = 14)

# Adjust the spacing at the bottom of the figure
plt.subplots_adjust(bottom=0)

# Save the figure to a file
plt.savefig('rank_assignment_distribution.png', bbox_inches='tight')

# Show the plot
plt.show()
```
